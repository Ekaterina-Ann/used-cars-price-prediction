```python
def pass_value_barh(df):
    try:
        (
            (df.isna().mean()*100)
            .to_frame()
            .rename(columns = {0: 'gap'})
            .query('gap > 0')
            .sort_values(by='gap', ascending=True)
            .plot(kind='barh', figsize=(15, 5), grid=True, legend=False, fontsize=16)
            .set_title('–ü—Ä–æ—Ü–µ–Ω—Ç –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –∫–∞–∂–¥–æ–º —Å—Ç–æ–ª–±—Ü–µ', fontsize=20)
        );
    except:
        print('–ü—Ä–æ–ø—É—Å–∫–æ–≤ –Ω–µ –æ—Å—Ç–∞–ª–æ—Å—å –∏–ª–∏ –ø—Ä–æ–∏–∑–æ—à–ª–∞ –æ—à–∏–±–∫–∞ –≤ –ø–µ—Ä–≤–æ–π —á–∞—Å—Ç–∏ —Ñ—É–Ω–∫—Ü–∏–∏ ')
```

```python
pass_value_barh(data)
```

```python
def clean(string):
    if ',' in string:
        list_str = string.split(',')
        for index, item in enumerate(list_str):
            list_str[index] = item.strip()
            result = '/'.join(list_str)
    elif '-' in string:
        list_str = string.split('-')
        for index, item in enumerate(list_str):
            list_str[index] = item.strip()
            result = '/'.join(list_str)
    else:
        result = string
    return result
```

```python
data['production_country'] = data['production_country'].apply(lambda x: clean(x))
```

```python
# –í—Ç–æ—Ä–æ–π –≤–∞—Ä–∏–∞–Ω—Ç —Ä–µ—à–µ–Ω–∏—è
df_list = []
def clean_duplicates(serrie):
    name = serrie.name
    before = len(serrie.unique())
    serrie = serrie.str.lower().str.strip()
    after = len(serrie.unique())
    df_list.append([name, before, after])
    return serrie
```

```python
def categorize(name):
    return name.split(',')[0]
```

```python
data['first_director'] = data['director'].apply(categorize)
data['first_genre'] = data['genres'].apply(categorize)
data.head()

# data[['first_director', 'first_genre']] = data[['director', 'genres']].apply(categorize)
```

- –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ, –∫–∞–∫—É—é –¥–æ–ª—é –æ—Ç –æ–±—â–µ–≥–æ –±—é–¥–∂–µ—Ç–∞ —Ñ–∏–ª—å–º–∞ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç –≥–æ—Å—É–¥–∞—Ä—Å—Ç–≤–µ–Ω–Ω–∞—è –ø–æ–¥–¥–µ—Ä–∂–∫–∞.

```python
def share(row):
    rf = row['refundable_support']
    nrf = row['nonrefundable_support']
    b = row['budget']
    try:
        result = round((rf + nrf) / b, 2)
    except ZeroDivisionError:
        result = 0
    return result
```

```python
data['support_share'] = data.apply(share, axis = 1)
```

#### –ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å —Å–±–æ—Ä–æ–≤ —Ñ–∏–ª—å–º–∞ –æ—Ç –≤–æ–∑—Ä–∞—Å—Ç–Ω–æ–≥–æ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è.

- –û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, –≤–ª–∏—è–µ—Ç –ª–∏ –≤–æ–∑—Ä–∞—Å—Ç–Ω–æ–µ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ –∞—É–¥–∏—Ç–æ—Ä–∏–∏ (¬´6+¬ª, ¬´12+¬ª, ¬´16+¬ª, ¬´18+¬ª –∏ —Ç. –¥.) –Ω–∞ —Å–±–æ—Ä—ã —Ñ–∏–ª—å–º–∞ –≤ –ø—Ä–æ–∫–∞—Ç–µ –≤ –ø–µ—Ä–∏–æ–¥ —Å 2015 –ø–æ 2019 –≥–æ–¥? –§–∏–ª—å–º—ã —Å –∫–∞–∫–∏–º –≤–æ–∑—Ä–∞—Å—Ç–Ω—ã–º –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ–º —Å–æ–±—Ä–∞–ª–∏ –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ –¥–µ–Ω–µ–≥ –≤ –ø—Ä–æ–∫–∞—Ç–µ? –ú–µ–Ω—è–µ—Ç—Å—è –ª–∏ –∫–∞—Ä—Ç–∏–Ω–∞ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –≥–æ–¥–∞? –ï—Å–ª–∏ –¥–∞, –ø—Ä–µ–¥–ø–æ–ª–æ–∂–∏—Ç–µ, —Å —á–µ–º —ç—Ç–æ –º–æ–∂–µ—Ç –±—ã—Ç—å —Å–≤—è–∑–∞–Ω–æ.

–ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –æ–±—â–∏–µ —Å–±–æ—Ä—ã —Ñ–∏–ª—å–º–æ–≤ —Å —Ä–∞–∑–Ω—ã–º –≤–æ–∑—Ä–∞—Å—Ç–Ω—ã–º –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ–º. –ü—Ä–∏–º–µ–Ω–∏–º —Ç–∞–∫–∂–µ —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏—é.

```python
# data_filtered = data.query('1e6 <= box_office <= 1.2e9 and 2015 <= year')
data_filtered = data[(data['year'] >= 2015) & (data['box_office'] <= 1.2e9) & (1e6 <= data['box_office'])]
```

```python
(
    data_filtered
    .pivot_table(index='age_restriction', values='box_office', aggfunc='sum')
    .plot(kind='bar', figsize=(15, 5), alpha=0.5, edgecolor='black')
)
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(['–°—É–º–º–∞ –∫–∞—Å—Å–æ–≤—ã—Ö —Å–±–æ—Ä–æ–≤'] )
plt.title('–°—É–º–º–∞ –∫–∞—Å—Å–æ–≤—ã—Ö —Å–±–æ—Ä–æ–≤ —Ñ–∏–ª—å–º–æ–≤ —Å —Ä–∞–∑–Ω—ã–º –≤–æ–∑—Ä–∞—Å—Ç–Ω—ã–º –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ–º' + '\n', fontsize=20)
plt.xlabel('–í–æ–∑—Ä–∞—Å—Ç–Ω–æ–µ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ', fontsize=14)
plt.ylabel('–°—É–º–º–∞ –∫–∞—Å—Å–æ–≤—ã—Ö —Å–±–æ—Ä–æ–≤', fontsize=14);
```

–ú—ã –≤–∏–¥–∏–º, —á—Ç–æ –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ –¥–µ–Ω–µ–≥ –≤ –ø—Ä–æ–∫–∞—Ç–µ —Å–æ–±—Ä–∞–ª–∏ —Ñ–∏–ª—å–º—ã —Å –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ–º 16+. –ê –≤–æ—Ç –º–µ–Ω—å—à–µ –≤—Å–µ–≥–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è 0+. –°—Ç–æ–∏—Ç –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã—à–µ–¥—à–∏—Ö —Ñ–∏–ª—å–º–æ–≤ –≤ –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏.

```python
(
    data_filtered
    .pivot_table(index='age_restriction', values='puNumber', aggfunc='count')
    .plot(kind='bar', figsize=(15, 5), alpha=0.5, edgecolor='black') #, subplots=True
)
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(['–°—É–º–º–∞ –∫–∞—Å—Å–æ–≤—ã—Ö —Å–±–æ—Ä–æ–≤'] )
plt.title('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ñ–∏–ª—å–º–æ–≤ —Å —Ä–∞–∑–Ω—ã–º –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ–º' + '\n', fontsize=20)
plt.xlabel('–í–æ–∑—Ä–∞—Å—Ç–Ω–æ–µ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ', fontsize=14)
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ñ–∏–ª—å–º–æ–≤', fontsize=14);
```

–î–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ —Ñ–∏–ª—å–º–æ–≤ —Å –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ–º 0+ –≤—ã—à–ª–æ –º–µ–Ω—å—à–µ –≤—Å–µ–≥–æ. –û–¥–Ω–∞–∫–æ –Ω–∞–∏–±–æ–ª–µ–µ –º–Ω–æ–≥–æ—á–∏—Å–ª–µ–Ω–∞ –∫–∞—Ç–µ–≥–æ—Ä–∏—è 18+, –∞ 16+ –Ω–∞ –≤—Ç–æ—Ä–æ–º –º–µ—Å—Ç–µ. –¢–µ–º –Ω–µ –º–µ–Ω–µ–µ —Ñ–∏–ª—å–º—ã 16+ —Å–æ–±—Ä–∞–ª–∏ –±–æ–ª—å—à–µ –¥–µ–Ω–µ–≥. –ü–æ—Å–º–æ—Ç—Ä–∏–º –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã: –º–µ–¥–∏–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å–±–æ—Ä–æ–≤.

- –ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –≤–∑–∞–∏–º–æ—Å–≤—è–∑—å –≥–æ—Å–ø–æ–¥–¥–µ—Ä–∂–∫–∏ –∏ –æ—Ü–µ–Ω–∫–∏ –∑—Ä–∏—Ç–µ–ª—è–º–∏ —Ñ–∏–ª—å–º–∞.

–°—Ç–æ–∏—Ç —Ä–∞–∑–¥–µ–ª–∏—Ç—å –¥–∞–Ω–Ω—ã–µ –ø–æ –≥–æ–¥–∞–º, —Ç–∞–∫ –∫–∞–∫ —Å—Ç—Ä–∞—Ç–µ–≥–∏—è —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏—è —Ç–∞–∫–∂–µ –º–æ–∂–µ—Ç –ø—Ä–∏—Ç–µ—Ä–ø–µ–≤–∞—Ç—å –∏–∑–º–µ–Ω–µ–Ω–∏—è —Å–æ –≤—Ä–µ–º–µ–Ω–µ–º.

```python
budget_films_by_ratings = (
    data_budget
    .query('year > 2014')
    .pivot_table(index='year', values='ratings', aggfunc='median')
    .rename(columns = {'ratings': '–†–µ–π—Ç–∏–Ω–≥_—Ñ–∏–ª—å–º–æ–≤_—Å_–≥–æ—Å–ø–æ–¥–¥–µ—Ä–∂–∫–æ–π'})
    #.sort_values(by='–†–µ–π—Ç–∏–Ω–≥_—Ñ–∏–ª—å–º–æ–≤_—Å_–≥–æ—Å–ø–æ–¥–¥–µ—Ä–∂–∫–æ–π', ascending=False)
).join((
    data.query('year > 2014 and budget.isna()').pivot_table(index='year', values='ratings', aggfunc='median')
        .rename(columns = {'ratings': '–†–µ–π—Ç–∏–Ω–≥_—Ñ–∏–ª—å–º–æ–≤'})
))
budget_films_by_ratings
```

–†–µ–π—Ç–∏–Ω–≥–∏ —Ñ–∏–ª—å–º–æ–≤ —Å –≥–æ—Å–ø–æ–¥–¥–µ—Ä–∂–∫–æ–π –≤ —Ü–µ–ª–æ–º –Ω–∏–∂–µ —á–µ–º –¥–ª—è –≤—Å–µ—Ö –æ—Å—Ç–∞–ª—å–Ω—ã—Ö. –°—Ç–æ–∏—Ç –ø—Ä–æ–≤–µ—Ä–∏—Ç—å —Ç–∞–∫–∂–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–∞–∫–∏—Ö —Ñ–∏–ª—å–º–æ–≤ –ø–æ –≥–æ–¥–∞–º, —Ç–∞–∫ –∫–∞–∫ –µ—Å–ª–∏ –æ—Ü–µ–Ω–æ–∫ –∑–∞ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã–π –≥–æ–¥ –º–∞–ª–æ, —Ç–æ –æ–Ω–∏ –Ω–µ —è–≤–ª—è—é—Ç—Å—è –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω—ã–º–∏.

–ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ñ–∏–ª—å–º–æ–≤ —Å –≥–æ—Å–ø–æ–¥–¥–µ—Ä–∂–∫–æ–π –ø–æ –≥–æ–¥–∞–º.

```python
(
    data_budget
    .query('year > 2014')
    .pivot_table(index='year', values='puNumber', aggfunc='count')
    .rename(columns = {'puNumber': '–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ_—Ñ–∏–ª—å–º–æ–≤_—Å_–≥–æ—Å–ø–æ–¥–¥–µ—Ä–∂–∫–æ–π'})
    #.sort_values(by='–†–µ–π—Ç–∏–Ω–≥_—Ñ–∏–ª—å–º–æ–≤_—Å_–≥–æ—Å–ø–æ–¥–¥–µ—Ä–∂–∫–æ–π', ascending=False)
).join((
    data.query('year > 2014').pivot_table(index='year', values='puNumber', aggfunc='count')
        .rename(columns = {'puNumber': '–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ_—Ñ–∏–ª—å–º–æ–≤'})
))
```

–í —Å—Ä–µ–¥–Ω–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–æ —Ñ–∏–ª—å–º–æ–≤ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ–µ.

```python
budget_films_by_ratings.plot(grid=(True), figsize=(15, 5))
plt.grid(True)
plt.xticks(rotation=0)
plt.legend(loc='lower right')
plt.title('–†–µ–π—Ç–∏–Ω–≥ —Ñ–∏–ª—å–º–æ–≤ –ø–æ –≥–æ–¥–∞–º' + '\n', fontsize=20)
plt.xlabel('–ì–æ–¥', fontsize=14)
plt.ylabel('–†–µ–π—Ç–∏–Ω–≥ —Ñ–∏–ª—å–º–æ–≤', fontsize=14);
```

–í–∏–¥–∏–º, —á—Ç–æ –≤ —Å—Ä–µ–¥–Ω–µ–º —Ä–µ–π—Ç–∏–Ω–≥ —Ñ–∏–ª—å–º–æ–≤ —Å –≥–æ—Å–ø–æ–¥–¥–µ—Ä–∂–∫–æ–π —á—É—Ç—å –Ω–∏–∂–µ —á–µ–º —É –≤—Å–µ—Ö –æ—Å—Ç–∞–ª—å–Ω—ã—Ö —Ñ–∏–ª—å–º–æ–≤ –∑–∞ —Ç–µ –∂–µ –≥–æ–¥—ã.

- –°—Ä–∞–≤–Ω–∏–º —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–æ —Ä–∞–∑–Ω—ã–º –∂–∞–Ω—Ä–∞–º –º–µ–∂–¥—É —Ñ–∏–ª—å–º–∞–º–∏ —Å –≥–æ—Å–ø–æ–¥–¥–µ—Ä–∂–∫–æ–π –∏ –±–µ–∑ –Ω–µ—ë.

```python
budget_films_by_ratings = (
    data_budget
    .pivot_table(index='first_genre', values='ratings', aggfunc='median')
    .rename(columns = {'ratings': '–†–µ–π—Ç–∏–Ω–≥_—Ñ–∏–ª—å–º–æ–≤_—Å_–≥–æ—Å–ø–æ–¥–¥–µ—Ä–∂–∫–æ–π'})
    #.sort_values(by='–†–µ–π—Ç–∏–Ω–≥_—Ñ–∏–ª—å–º–æ–≤_—Å_–≥–æ—Å–ø–æ–¥–¥–µ—Ä–∂–∫–æ–π', ascending=False)
).join((
    data.query('budget.isna()').pivot_table(index='first_genre', values='ratings', aggfunc='median')
        .rename(columns = {'ratings': '–†–µ–π—Ç–∏–Ω–≥_—Ñ–∏–ª—å–º–æ–≤'})
))
budget_films_by_ratings
```

–£–¥–∞–ª–∏–º –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω—É—é —Å—Ç—Ä–æ–∫—É —Å –Ω–µ–ø–æ–Ω—è—Ç–Ω—ã–º –∂–∞–Ω—Ä–æ–º '—Ä–µ–∞–ª—å–Ω–æ–µ —Ç–≤', —Ç–∞–∫ –∫–∞–∫ –æ–Ω–∞ –Ω–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–∞.

```python
budget_films_by_ratings = budget_films_by_ratings[budget_films_by_ratings.index != '—Ä–µ–∞–ª—å–Ω–æ–µ —Ç–≤']
budget_films_by_ratings
```

```python
budget_films_by_ratings.plot(kind='bar', grid=(True), figsize=(15, 5))
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(loc='upper left')
plt.title('–†–µ–π—Ç–∏–Ω–≥–∏ —Ñ–∏–ª—å–º–æ–≤ —Å –≥–æ—Å–ø–µ–¥–¥–µ—Ä–∂–∫–æ–π –∏ –±–µ–∑' + '\n', fontsize=20)
plt.xlabel('–ñ–∞–Ω—Ä', fontsize=14)
plt.ylabel('–†–µ–π—Ç–∏–Ω–≥', fontsize=14);
```

–°—É–¥—è –ø–æ –≥—Ä–∞—Ñ–∏–∫—É —á–∞—â–µ —Ä–µ–π—Ç–∏–Ω–≥ —Ñ–∏–ª—å–º–æ–≤ —Å –≥–æ—Å–ø–æ–¥–¥–µ—Ä–∂–∫–æ–π —á—É—Ç—å –Ω–∏–∂–µ —á–µ–º –≤ —Å—Ä–µ–¥–Ω–µ–º, –∑–∞ –∏—Å–∫–ª—é—á–µ–Ω–∏–µ–º —Ñ–∏–ª—å–º–æ–≤ –≤ –∫—Ä–∏–º–∏–Ω–∞–ª—å–Ω–æ–º –∂–∞–Ω—Ä–µ, –∏ –≤ –∂–∞–Ω—Ä–µ —Ñ–µ–Ω—Ç–µ–∑–∏.

- –ü–æ—Å–º–æ—Ç—Ä–∏–º –∫–∞–∫–∏–µ —Ç–∏–ø—ã —Ñ–∏–ª—å–º–æ–≤ —á–∞—â–µ —Ñ–∏–Ω–∞–Ω—Å–∏—Ä—É—é—Ç—Å—è.

```python
budget_films_by_genre = (
    data_budget
    .pivot_table(index='first_genre', values='puNumber', aggfunc='count')
    .rename(columns = {'puNumber': '–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ_—Ñ–∏–ª—å–º–æ–≤_—Å_–≥–æ—Å–ø–æ–¥–¥–µ—Ä–∂–∫–æ–π'})
    .sort_values(by='–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ_—Ñ–∏–ª—å–º–æ–≤_—Å_–≥–æ—Å–ø–æ–¥–¥–µ—Ä–∂–∫–æ–π', ascending=False)
).join((
    data.pivot_table(index='first_genre', values='puNumber', aggfunc='count')
        .rename(columns = {'puNumber': '–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ_—Ñ–∏–ª—å–º–æ–≤_–≤—Å–µ–≥–æ'})
))
budget_films_by_genre.drop(budget_films_by_genre.tail(1).index,inplace=True)
budget_films_by_genre
```
