```python
def pass_value_barh(df):
    try:
        (
            (df.isna().mean()*100)
            .to_frame()
            .rename(columns = {0: 'gap'})
            .query('gap > 0')
            .sort_values(by='gap', ascending=True)
            .plot(kind='barh', figsize=(15, 5), grid=True, legend=False, fontsize=16)
            .set_title('Процент пропусков в каждом столбце', fontsize=20)
        );
    except:
        print('Пропусков не осталось или произошла ошибка в первой части функции ')
```

##### Обработаем дубликаты в столбце production_country

Для начала приведём всё к нижнему регистру и проверим на очевидные дубликаты.

Посмотрим на уникальные значения столбца `production_country`. Видно, что одни и те же страны расписаны разным способом. Например: "китай" - "кнр", "казахстан" - "республика казахстан", "новая зеландия" - "н.зеландия", "англия" - "великобритания", "ю.корея","южная корея", "республика корея" - "корея", "республика беларусь" - "беларусь" и т.д.

Также для строк, где указано несколько стран есть разные форматы записи: через запятую, через тире. Межд ними могут быть и не быть пробелы.

Приведём всё к единому формату, и разделим страны "/", удалим лишние пробелы перед и после названий,  при этом сохраним пробелы в написании таких стран как "новая зеландия".

```python
# data['production_country'] = data['production_country'].replace(['-', ','], '/', regex=True)
# data['production_country'] = data['production_country'].replace([' /', '/ ', '/  ', '/   '], '/', regex=True)
```

```python
len(data['production_country'].unique())
```

```python
def clean(string):
    if ',' in string:
        list_str = string.split(',')
        for index, item in enumerate(list_str):
            list_str[index] = item.strip()
            result = '/'.join(list_str)
    elif '-' in string:
        list_str = string.split('-')
        for index, item in enumerate(list_str):
            list_str[index] = item.strip()
            result = '/'.join(list_str)
    else:
        result = string
    return result
```

```python
data['production_country'] = data['production_country'].apply(lambda x: clean(x))
```

```python
len(data['production_country'].unique())
```

```python
data[['title', 'production_country']].duplicated().sum()
```

Видим, что уникальных значений стало меньше. Но есть случаи где одни и те же страны расположены в разном порядке. Конечно это может быть связано с тем, что какя-то из них являлась главной в конкретном фильме, но для нас это неважно. Лучше чтобы все страны были записаны в алфавитном порядке. Это поможет избежать неявные дубликатов.

```python
def arrange_names(string):
    list_str = string.split('/')
    string = '/'.join(sorted(list_str))
    return string
```

```python
data['production_country'] = data['production_country'].apply(lambda x: arrange_names(x))
```

```python
len(data['production_country'].unique())
```

```python
data[['title', 'production_country']].duplicated().sum()
```

```python
data['production_country'].unique()
```

- Посчитайте, какую долю от общего бюджета фильма составляет государственная поддержка.

```python
def share(row):
    rf = row['refundable_support']
    nrf = row['nonrefundable_support']
    b = row['budget']
    try:
        result = round((rf + nrf) / b, 2)
    except ZeroDivisionError:
        result = 0
    return result
```

```python
data['support_share'] = data.apply(share, axis = 1)
```
